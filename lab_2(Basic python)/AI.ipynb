## 1) Змінні та типи

````
a = 3
b = 2.5
flag = True
text = "Python"
(type(a), type(b), type(flag), type(text))
````

**Очікуваний вивід:** `(<class 'int'>, <class 'float'>, <class 'bool'>, <class 'str'>)`

## 2) Рядки та f‑рядки

````
s = "Python"
print((s[0], s[-1], s[1:4], s.upper()))
````

**Очікуваний вивід:** `('P', 'n', 'yth', 'PYTHON')`

````
name = "Nazar"
print(f"Привіт, {name}! {2+2=}")
````

**Очікуваний вивід:** `Привіт, Nazar! 2+2=4`

## 3) Колекції: `list` / `tuple` / `set` / `dict`

````
nums = [1, 3, 5]
nums.append(7)
print(nums, nums[:2])
````

**Очікуваний вивід:** `[1, 3, 5, 7] [1, 3]`

````
point = (10, 20)
print(point[0])
````

**Очікуваний вивід:** `10`

````
uniq = set([1, 2, 2, 3])
print(uniq)
````

**Очікуваний вивід:** `{1, 2, 3}` (порядок у множин може відрізнятись)

````
rates = {"usd": 39, "eur": 42}
print(rates["usd"], rates.get("gbp", 0))
````

**Очікуваний вивід:** `39 0`

## 4) Розгалуження і цикли (включно з comprehension)

````
x = 0
if x < 0:
    msg = "негативне"
elif x == 0:
    msg = "нуль"
else:
    msg = "позитивне"
print(msg)
````

**Очікуваний вивід:** `нуль`

````
squares = [n*n for n in range(5)]
evens = [n for n in range(1, 7) if n % 2 == 0]
print(squares, evens)
````

**Очікуваний вивід:** `[0, 1, 4, 9, 16] [2, 4, 6]`

## 5) Функції і `lambda`

````
def greet(name="World"):
    """Повертає вітання."""
    return f"Hello, {name}!"
print(greet("Nazar"), greet())
````

**Очікуваний вивід:** `Hello, Nazar! Hello, World!`

````
def stats(nums):
    mn, mx = min(nums), max(nums)
    return mn, mx, sum(nums)/len(nums)
print(stats([1, 2, 8]))
````

**Очікуваний вивід:** `(1, 8, 3.6666666666666665)`

````
print(sorted(["aa", "b", "cccc"], key=lambda s: len(s)))
````

**Очікуваний вивід:** `['b', 'aa', 'cccc']`

## 6) Обробка помилок (`try/except/finally`)

````
try:
    10 / 0
except ZeroDivisionError as e:
    print("Помилка:", e)
finally:
    print("Готово")
````

**Очікуваний вивід:**

```
Помилка: division by zero
Готово
```

## 7) Контекст‑менеджер `with open(...)`

````
# запис
with open("demo.txt", "w", encoding="utf-8") as f:
    print("перша", file=f)
    print("друга", file=f)

# читання
with open("demo.txt", encoding="utf-8") as f:
    print(f.read())
````

**Очікуваний вивід:**

```
перша
друга
```

## 8) Імпорт модулів

````
import math
from statistics import mean
print(math.isclose(math.pi, 3.14159, rel_tol=1e-5), mean([1, 2, 9]))
````

**Очікуваний вивід:** `True 4`
