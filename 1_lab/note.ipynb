## ChatGPT:

````python
#!/usr/bin/env python3
"""
first_program.py — Перша навчальна програма «Фібоначчі з поясненням»

Ідея: обчислити перші N чисел Фібоначчі.
Демонструє: функції, перевірку помилок, типи, докстрінги, argparse, юзабельний CLI.
"""
from __future__ import annotations
import argparse
from typing import List

def fibonacci(n: int) -> List[int]:
    """
    Повертає список із перших n чисел Фібоначчі.

    Валідація:
      - n має бути цілим і 0 <= n <= 10_000 (щоб не вбити ПК величезним виводом).
    """
    if not isinstance(n, int):
        raise TypeError("n має бути int")
    if n < 0:
        raise ValueError("n має бути невід'ємним")
    if n > 10_000:
        raise ValueError("n занадто велике (макс 10000)")

    a, b = 0, 1
    out: List[int] = []
    for _ in range(n):
        out.append(a)
        a, b = b, a + b
    return out

def main(argv: list[str] | None = None) -> int:
    parser = argparse.ArgumentParser(
        description="Вивести перші N чисел Фібоначчі."
    )
    parser.add_argument(
        "-n", "--count", type=int, default=10,
        help="Скільки чисел вивести (за замовчуванням: 10)"
    )
    args = parser.parse_args(argv)
    seq = fibonacci(args.count)
    print(" ".join(map(str, seq)))
    return 0

if __name__ == "__main__":
    raise SystemExit(main())
````
